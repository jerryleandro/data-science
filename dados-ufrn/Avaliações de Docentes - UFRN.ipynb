```python
#Importes

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Lendo o dados no arquivo CVC e exibindo

avaliacoes = pd.read_csv(('files/avaliacaodocencia.csv'), sep=';')
avaliacoes.head(5)
```

```python
# Pegando apenas das avaliações a partir de 2017
# Professores com melhores avaliações
avaliacao_2017 = avaliacoes[avaliacoes.ano == 2017]

#Melhores avaliacoes do ano de 2017
melhores_avaliacoes = avaliacao_2017[avaliacao_2017.postura_profissional_media > 9.9]

#Melhores avaliações do ano de 2017 cujo a quantidade de alunos foram maior que 50

melhores_avaliacoes_maior = avaliacao_2017[avaliacao_2017.qtd_discentes > 120]

melhores_avaliacoes_maior.head(5)
```

```python
# Gráfico de barras informando as avaliações por professor que tem nota maior que 9.9 e mais de 120 alunos
sns.set(font_scale=1.4)
grafico = sns.factorplot("nome_docente", data=melhores_avaliacoes_maior, kind="count",size=5, aspect=1.5)
grafico.set_xticklabels(rotation=45)
plt.title('Avaliação dos Docentes')
plt.xlabel('')
plt.ylabel('Nota')
plt.show()
```
